```python
import requests
from bs4 import BeautifulSoup 
import re
import jieba.posseg as pseg
import jieba
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
import jieba.analyse
from wordcloud import WordCloud
```

## 新闻爬虫和中文清洗

```python
# 请求url
url = 'https://3w.huanqiu.com/a/c36dc8/41hhhzLG1zl'
html = requests.get(url, timeout = 10)
content = html.content
##print(content)
```

```python
# 通过content创建beautifulsoup对象
soup = BeautifulSoup(content,'html.parser',from_encoding = 'utf-8')
text = soup.get_text()
print(text)
```

```python
import jieba.posseg as pseg
```

```python
#  获取人物.地点
words = pseg.lcut(text)
# 人物集合
news_person = {word for word,flag in words if flag=='nr'}
news_place = {word for word,flag in words if flag=='ns'}
print('新闻中的人物:', news_person)
print('新闻中的地点:', news_place)
```

```python
# 提取中文以及相关的标点符号
r1 = "[a-zA-Z0-9\s+\.\!\/_$%^*(+\"\']+|[+——！？、：；;《》“”~@#￥%……&*（）=-{::::})|【】<<]+"
text = re.sub(r1,'',text)
print(text)
```

## 获取文章关键词

```python
# 从文本中获取词
words = jieba.lcut(text)
print(words)
```

```python
vectorizer = CountVectorizer()
count = vectorizer.fit_transform(words)
print("count=", count)
```

```python
## 计算TFIDF
transformer = TfidfTransformer()
## TFIDF矩阵
tfidf_matrix= transformer.fit_transform(count)
print(tfidf_matrix.toarray())
```

```python
## TfidfVectorizer可以把CountVecterizer和Tfidftransformer合并起来
tfidf_vec = TfidfVectorizer()
tfidf_matrix = tfidf_vec.fit_transform(words)
print(tfidf_matrix.toarray())
### 获得文本的关键词
print(tfidf_vec.get_feature_names())
print(tfidf_vec.vocabulary_)
```

```python
### 或者使用jieba中的textrank得到关键词
jieba.analyse.textrank(text, topK=20, withWeight=True, allowPOS=('ns', 'n', 'vn', 'v'))
```

```python
from textrank4zh import TextRank4Keyword, TextRank4Sentence
# 输出关键词，设置文本小写，窗口为2
tr4w = TextRank4Keyword()
tr4w.analyze(text=text, lower=True, window=2)
print('关键词：')
for item in tr4w.get_keywords(20, word_min_len=1):
    print(item.word, item.weight)
```

## 新闻关键句摘要

```python
# 输出重要的句子
tr4s = TextRank4Sentence()
tr4s.analyze(text=text, lower=True, source = 'all_filters')
print('摘要：')
# 重要性较高的三个句子
for item in tr4s.get_key_sentences(num=3):
    # index是语句在文本中位置，weight表示权重
    print(item.index, item.weight, item.sentence)
```

## 词云可视化

```python
# 去掉停用词
def remove_stop_words(f):
    # 获取停用词list 
    stop_words = [line.strip() for line in open('./cn_stopwords.txt').readlines()]
    for stop_word in stop_words:
        f = f.replace(stop_word,'')
    return f
```

```python
## 可视化
def create_word_cloud(f):
    f = remove_stop_words(f)
    seg_list = jieba.lcut(f)
    cut_text = ' '.join(seg_list)
    wc = WordCloud(
        max_words=100,
        width=2000,
        height=1200,
        font_path='./msyh.ttf'
    )
    wordcloud = wc.generate(cut_text)
    wordcloud.to_file("wordcloud.jpg")
    
create_word_cloud(text)
```
