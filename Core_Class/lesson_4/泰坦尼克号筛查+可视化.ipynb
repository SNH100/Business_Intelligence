```python
# step7
import seaborn as sns
import matplotlib.pyplot as plt
train.isnull().sum()
```

```python
train_nan_per = (train.isnull().sum() / len(train)) * 100
train_nan_per = train_nan_per.drop(train_nan_per[train_nan_per == 0].index).sort_values(ascending = False)
missing_data = pd.DataFrame({'Missing Ratio':train_nan_per})

sns.barplot(x = train_nan_per.index, y = train_nan_per)
plt.xlabel('Features', fontsize=15)
plt.ylabel('Percent of missing values', fontsize=15)
plt.title('Percent missing data by feature', fontsize=15)
```

```python
sns.relplot(x=train.index, y='Age', kind='line', data=train, ci=None)
plt.show()
```

```python
corr = train.corr()
sns.set(style='white')
f, ax = plt.subplots(figsize=(11, 9))
# 生成自定义的散色图
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# 为热力图设置长宽比
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
           square=True, linewidths=.5, cbar_kws={"shrink":.5})
plt.show()
```
