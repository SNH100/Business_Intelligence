```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
data_path = 'raw_data'

df_train = pd.read_csv(os.path.join(data_path, 'train.csv'))
df_test = pd.read_csv(os.path.join(data_path, 'test_noLabel.csv'))
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
df_test['Label'] = -1
df_feature = df_train.append(df_test).reset_index(drop=True)
df_feature.head()
```

```python
cat_col = [i for i in df_feature.select_dtypes(object).columns if i not in ['ID', 'Label']]
for i in tqdm(cat_col):
    lbl = LabelEncoder()
    df_feature[i] = lbl.fit_transform(df_feature[i].astype(str))
```

```python
df_feature.head()
```

```python
feats = [i for i in df_feature.columns if i not in ['ID', 'Label']]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
tar = df_feature[df_feature['Label']!=-1][feats]
y =df_feature[df_feature['Label']!=-1]['Label']
testx= df_feature[df_feature['Label']==-1][feats]

train_x, vali_x, train_y, vali_y = train_test_split(tar, y, test_size=0.3, random_state=42)
model.fit(train_x, train_y)
test_pre = model.predict_proba(testx)[:,1]
```

```python
#线下验证
vali_pre = model.predict_proba(vali_x)[:,1]
p = 0.3
vali_pre[vali_pre < p] = 0
vali_pre[vali_pre >= p] = 1
print('线下AUC:           '+str(roc_auc_score(vali_y, vali_pre)))
```

```python
pre = df_feature[df_feature['Label']==-1][['ID']]
pre['Label'] = test_pre
pre['Label']= (pre['Label'] >= p).astype(int)
os.makedirs('sub', exist_ok=True)
pre[['ID', 'Label']].to_csv('sub/baseline.csv',index=False )
```
